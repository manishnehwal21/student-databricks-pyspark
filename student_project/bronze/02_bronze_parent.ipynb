```python
from pyspark.sql import SparkSession
```

```python
from student_project.configs.db_config import (
    DB_HOST, DB_NAME, DB_USER, DB_PASSWORD
)

jdbc_url = f"jdbc:mysql://{DB_HOST}:3306/{DB_NAME}"

jdbc_props = {
    "user": DB_USER,
    "password": DB_PASSWORD,
}
```

```python
parent_df = spark.read.jdbc(
    url=jdbc_url,
    table="parent",
    properties=jdbc_props
)
```

```python
parent_df
```

```python
parent_df.show(5)
```

```python
print("Total parent records:", parent_df.count())
```

```python
parent_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("student_project.bronze_parent")
```
