```python
# Notebook: 03_bronze_class
# Layer: Bronze
# Source: MySQL - class table
# Target: Databricks Managed Delta Table
```

```python
from student_project.configs.db_config import (
    DB_HOST, DB_NAME, DB_USER, DB_PASSWORD
)

jdbc_url = f"jdbc:mysql://{DB_HOST}:3306/{DB_NAME}"

jdbc_props = {
    "user": DB_USER,
    "password": DB_PASSWORD,
}
```

```python
class_df = spark.read.jdbc(
    url=jdbc_url,
    table="class",
    properties=jdbc_props
)
```

```python
class_df
```

```python
class_df.show(5)
```

```python
print("Total class records:", class_df.count())
```

```python
class_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("student_project.bronze_class")
```
